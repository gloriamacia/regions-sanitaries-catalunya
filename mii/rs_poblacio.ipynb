# Càlcul de la prevalència i incidència de la MII per SS

```python
import pandas as pd
df = pd.read_csv('/content/ss_poblacio.csv', dtype={'codi SS': object})
df.head(3)
```

S'extreu d'aquest [article científic](https://www.tandfonline.com/doi/full/10.1080/07853890.2018.1523550) la prevalència i l'incidència de la MII a Catalunya. 

Screenshot 2021-12-26 at 18.47.11.png

* [2011-2016] The average incidence over this six-year period was 36.3 new cases of IBD per 100,000 inhabitants per year: 21.6 UC and 14.7 CD.
* The prevalence per 100,000 inhabitants in 2016 was 353.9 for UC and 191.4 for CD. 

```python
prevalence_rate = (353.9 + 191.4)/100000
incidence_rate = (21.6 + 14.7)/100000
df['prevalència'] = round(df['població oficial']*prevalence_rate)
df['incidència'] = round(df['població oficial']*incidence_rate)
df.head(2)
```

```python
df.to_csv('/content/prevalencia_incidencia_mii_per_ss.csv', index=False)
```
