Nota: Malgrat que no s'ha calculat la població a nivell d'AGA això també seria possible seguint el mateix procediment amb aquestes [dades](https://catsalut.gencat.cat/ca/proveidors-professionals/registres-catalegs/catalegs/territorials-unitats-proveidores/). 

```python
import pandas as pd
abs = pd.read_csv("/content/2021_registre_central_poblacio_per_abs_catsalut.csv")
f = {'codi Àrea Bàsica de Saut': 'first', 'Àrea Bàsica de Salut': 'first', 'població oficial': 'sum'}
abs = abs.groupby(['Àrea Bàsica de Salut'], as_index=False).agg(f)
abs.head(3)
```

```python
ss = pd.read_csv('/content/06-Cataleg ABS.csv')
ss.head(3)
```

```python
ss = ss.merge(abs, how='left', right_on='codi Àrea Bàsica de Saut', left_on='codi ABS')
ss.drop(labels=['codi Àrea Bàsica de Saut', 'Àrea Bàsica de Salut'], axis=1, inplace=True)
ss.head(3)
```

## Població per Àrea Bàsica de Salut (ABS)



```python
f = {'codi ABS': 'first', 'descripcio ABS': 'first', 'codi SS': 'first', 'descripcio SS': 'first', 'codi RS': 'first', 'descripcio RS': 'first', 'població oficial': 'mean'}
absp = ss.groupby(['codi ABS'], as_index=False).agg(f)
absp.to_csv('/content/abs_poblacio.csv', index=False)
absp.head(3)
```

## Població per Sector Sanitari (SS)


```python
f = {'codi SS': 'first', 'descripcio SS': 'first', 'descripcio SS': 'first', 'codi RS': 'first', 'descripcio RS': 'first', 'població oficial': 'sum'}
ssp = absp.groupby(['codi SS'], as_index=False).agg(f)
ssp.to_csv('/content/ss_poblacio.csv', index=False)
ssp.head(3)
```

## Població per Regió Sanitària (RS)

```python
# poblacio per regio sanitaria
f = {'codi RS': 'first', 'descripcio RS': 'first', 'població oficial': 'sum'}
rsp = ssp.groupby(['codi RS'], as_index=False).agg(f)
rsp.to_csv('/content/rs_poblacio.csv', index=False)
rsp.head(3)
```

## Població Total Catalunya (2021) en millions d'habitans

```python
rsp['població oficial'].sum() / 10**6
```
